```python
def get_data(zip_codes):
    restaurants = []
    params = {}


    for code in zip_codes:
        print(code)
        url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={zip_codes[code]}&radius=2000&type=restaurant&key={Google_API_Key}"
        response = requests.get(url, params = params)
        results =  json.loads(response.content)
        restaurants.extend(results['results'])
        time.sleep(2)

        while "next_page_token" in results:
            params['pagetoken'] = results['next_page_token'],
            response = requests.get(url, params = params)
            results = json.loads(response.content)
            restaurants.extend(results['results'])
            time.sleep(2)
        
    return restaurants
```

```python
def json_to_list(restaurants):
    rest_name = []
    loc_lat = []
    loc_long = []
    business_status = []
    place_id = []
    price_level = []
    rating = []
    total_ratings = []


    for i in range(len(restaurants)):
        rest = restaurants[i]
        
        try: 
            rest_name.append(rest['name'])
        except:
            rest_name.append('none')
        try: 
            loc_lat.append(rest['geometry']['location']['lat'])
            loc_long.append(rest['geometry']['location']['lng'])
        except:
            loc_lat.append('none')
            loc_long.append('none')
        try: 
            business_status.append(rest['business_status'])
        except: 
            business_status.append('none')
        try: 
            place_id.append(rest['place_id'])
        except:
            place_id.append('none')
        try: 
            price_level.append(rest['price_level'])
        except:
            price_level.append('none')
        try: 
            rating.append(rest['rating'])
        except:
            rating.append('none')
        try: 
            total_ratings.append(rest['user_ratings_total'])
        except: 
            total_ratings.append('none')
            
    return rest_name, loc_lat, loc_long, business_status, place_id, price_level, rating, total_ratings
```

```python
def get_place_data(place_id):
    dine_in = []
    addr = []
    url = []
    reviews = []
    fields = "dine_in,reviews,formatted_address,url"
    params = {}

    for place in place_id:
        place_url = f"https://maps.googleapis.com/maps/api/place/details/json?place_id={place}&fields={fields}&key={Google_API_Key}"

        response = requests.get(place_url, params = params)
        results =  json.loads(response.content)

        try: 
            dine_in.append(results['result']['dine_in'])
        except:
            dine_in.append('None')
        try:
            addr.append(results['result']['formatted_address'])
        except: 
            addr.append('None')
        try:
            url.append(results['result']['url'])
        except: 
            url.append('None')
        try:
            reviews.append(results['result']['reviews'])
        except: 
            reviews.append('None')   
    return dine_in, addr, url, reviews
```
